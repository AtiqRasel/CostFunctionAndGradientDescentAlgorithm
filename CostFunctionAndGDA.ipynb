```python

def gradient_descent(x,y):
  m=b=0
  learningRate=.00001
  n=len(x)
  iteration=10000
  previousCost=100000
  for i in range(iteration):
    ypredict=m*x+b
    costFunction=(1/n)*sum([val**2 for val in (y-ypredict)])
    dM=(-2/n)*sum(x*(y-ypredict))
    db=(-2/n)*sum(y-ypredict)
    m=m-learningRate*dM
    b=b-learningRate*db
    print("m{}, b{}, Cost {}".format(m,b,costFunction))  
    if math.isclose(costFunction,previousCost, abs_tol=1e-30):
      print("iteration {}".format(i))
      break
    else:
      previousCost=costFunction

  
```

```python
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
gradient_descent(x,y)
```
